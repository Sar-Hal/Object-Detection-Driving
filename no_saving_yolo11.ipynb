```python
import cv2
from ultralytics import YOLO

# Load YOLOv11 model (replace with the actual model file path)
model = YOLO("yolo11n.pt")  # Replace with the correct YOLOv11 model path if available

# Input: Single video filename
vid_name = input("Enter the video filename: ")

cap = cv2.VideoCapture(vid_name)

if not cap.isOpened():
    print(f"Error: Could not open video file {vid_name}.")
    exit()

# Create a resizable window
cv2.namedWindow("Real-time Object Detection", cv2.WINDOW_NORMAL)

print(f"Processing {vid_name}...")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLOv11 inference
    results = model(frame, conf=0.5)

    # Annotate frame
    annotated_frame = results[0].plot()

    # Show the annotated frame in a flexible window
    cv2.imshow("Real-time Object Detection", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()

print("Processing completed.")
```
